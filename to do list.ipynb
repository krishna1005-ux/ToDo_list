```python
from tkinter import *
from tkinter import messagebox
from tkcalendar import DateEntry

krish=Tk()
#krish.geometry("400x700")

label=Label(krish,text="üìù To-Do List")
label.pack()

#def submit():
    #print("you entered:", entrybox.get())
    #messagebox.showwarning("warning","no item selected to work")

def add():
    task=entrybox.get()
    date=due_date.get()

    if task.strip() == "":
        messagebox.showwarning("warning","please enter the task.")
        return
    full_task=f"{task} (Due: {date})"
    listbox.insert(END,full_task)
    entrybox.delete(0,END)

def delete():
    try:
        selected_item_index = listbox.curselection()[0]
        listbox.delete(selected_item_index)
    except IndexError:
        messagebox.showwarning("warning","no item selected to work")

def done():
    try:
        index = listbox.curselection()[0]
        task = listbox.get(index)
        if not task.startswith("‚úÖ "):
            listbox.delete(index)
            listbox.insert(index, "‚úÖ " + task)
        else:
            print("Task already marked as done.")
    except IndexError:
        messagebox.showwarning("warning","no item selected to work")

def save():
    with open("tasks.txt","w") as file:
        for i in range(listbox.size()):
            file.write(listbox.get(i) + "\n")
    messagebox.showinfo("success","task has been saved to text.txt")

def load():
    try:
        with open("tasks.txt","r") as file:
            listbox.delete(0,END)
            for line in file:
                listbox.insert(END,line.strip())
        messagebox.showinfo("success","task loaded from tasks.txt")
    except FileNotFoundError:
        messagebox.showwarning("Warning", "No saved task file found.")


listbox=Listbox(krish,fg="black",bg="#FFFFFF",font="arial 20",highlightthickness=2,highlightbackground="gray",highlightcolor="black" )
listbox.pack(anchor=CENTER)

label=Label(krish,text="ENTER NEW TASK",font="arial 12")
label.pack()

entrybox=Entry(krish,fg="black",bg="white",font="arial 20",highlightthickness=2,highlightbackground="gray",highlightcolor="black" )
entrybox.pack(pady=10)

#date picker label
date_label=Label(krish,text="Select Due Date: ",font="arial 12")
date_label.pack()

#date picker widget
due_date=DateEntry(krish,width=12,bg="#000400",fg="white",borderwidth=2)
due_date.pack()



#submit=Button(krish,text="submit",command=submit)
#submit.pack()

frame=Frame(krish)
frame.pack(pady=5)

add=Button(krish,text="add",command=add)
add.pack(side=LEFT,pady=5)

delete=Button(krish,text="delete",command=delete)
delete.pack(side=LEFT,pady=5)

mark=Button(krish,text="mark as done",command=done)
mark.pack(side=LEFT,pady=5)

#save button
save_data=Button(krish,text="Save Tasks",command=save)
save_data.pack(side=RIGHT,pady=5)

load_data=Button(krish,text="Load Tasks",command=load)
load_data.pack(side=BOTTOM,pady=5)

#save_data=Button(krish,text="Save Tasks",command=save)
#save_data.pack(side=LEFT,pady=5)

#load_data=Button(krish,text="Load Tasks",command=load)
#load_data.pack(side=LEFT,pady=5)


krish.mainloop()
```
